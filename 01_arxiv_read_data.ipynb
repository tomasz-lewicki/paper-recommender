```python
import pandas as pd
import json
```

```python
def line_generator(fname='arxiv-metadata-oai-snapshot.json'):
    with open(fname, 'r') as f:
        for line in f:
            yield line
```

```python
def make_dataframe(n=1000):
    li = []
    gen = line_generator()
    for idx, line in enumerate(gen):
        d = json.loads(line)
        li.append(d)        
        if idx >= n-1: break
    
    return pd.DataFrame(li)
```

```python
df = make_dataframe(10000)
```
